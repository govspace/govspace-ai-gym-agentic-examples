## Step 2: Basic Usage — Ask a Question

The syntax is:
```
%%ai <provider>:<model>
Your prompt here
```

Let's try it with Claude:

````python
%%ai anthropic:claude-sonnet-4-20250514
What are the three most important principles for designing safe AI agents? Keep it concise — one sentence each.
````

And with GPT:

````python
%%ai openai-chat:gpt-4o-mini
What are the three most important principles for designing safe AI agents? Keep it concise — one sentence each.
````

## Step 3: Output Formats

Use the `-f` flag to control how the output is rendered:

| Flag | Format | Use case |
|------|--------|----------|
| `-f text` | Plain text | Default |
| `-f markdown` | Rendered Markdown | Rich formatting, tables |
| `-f code` | Syntax-highlighted code | Generate code snippets |
| `-f math` | LaTeX math | Equations and formulas |
| `-f json` | Formatted JSON | Structured data |

````python
%%ai anthropic:claude-sonnet-4-20250514 -f code
Write a Python function that validates an email address using regex. Include docstring and type hints.
````

````python
%%ai anthropic:claude-sonnet-4-20250514 -f markdown
Create a comparison table of LangChain vs LangGraph vs raw SDK calls. Columns: Approach, Best For, Complexity, Flexibility.
````

## Step 4: Work With Your Variables

Use `{variable}` in your prompt to inject Python variables from your kernel. This is powerful for analyzing data on the fly.

````python
# Create some sample data
sample_data = """
Country,RegTech_Adoption_Score,AI_Policy_Maturity,Population_M
Singapore,92,High,5.9
UK,88,High,67.7
Canada,76,Medium,38.9
Kenya,61,Medium,54.0
Brazil,55,Low,214.3
""".strip()
````

````python
%%ai anthropic:claude-sonnet-4-20250514 -f markdown
Analyze this dataset and identify interesting patterns. What policy recommendations would you make?

{sample_data}
````

```python

# With format flag
%%ai anthropic:claude-sonnet-4-20250514 -f code

%%ai anthropic:claude-sonnet-4-20250514 -f markdown
Analyze this: {my_variable}

# Line magic for quick one-liners
%ai anthropic:claude-sonnet-4-20250514 Explain gradient descent in one sentence
```
